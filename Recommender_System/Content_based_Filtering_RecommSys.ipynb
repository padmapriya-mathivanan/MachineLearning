<h1> 7. List Similar Playlist using Cosine Similarity function


<h1> 
Listing 10 similar playlist 

```python

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=RuntimeWarning)

def getSimilarplaylist(rowNo):
    
    csResultL=[]
    csIndexL=[]
    
    
    rowNoVec=merged.loc[rowNo, 'VIBE':].values 
    # rowno is the parameter, rowNovec is the array list 
   
    for i in range(spaceSize): #
        iVec=merged.loc[i,'VIBE':].values
        #cosine similarity calculation
        csResultL.append(np.dot(rowNoVec, iVec)/(np.linalg.norm(rowNoVec)*np.linalg.norm(iVec)))
        #each row in dfMerged calculate cosine similarity with simulated user preference
        #which is produced by a random list . In this case manually input as 1231 
        csIndexL.append(i)
    
    dfResult=pd.DataFrame({'Cosine Similarity': csResultL})
    #csResultL is a list of all calculated cosine similarity for all rows compared against preference.
    #dfResult is dataframe with index same as row number of dfMerged and each row is csResultL's element
    print("Cosine similarity calculated for simulated user preference and recommended playlist")
    dfResult=dfResult.sort_values('Cosine Similarity', ascending=False).head(10)
    return dfResult

dfResult = getSimilarplaylist(26582)
print (dfResult)
indexList=dfResult.index.tolist()


print('The following are the 10 recommendations:')
pd.set_option('display.max_colwidth',-1)
dfDisplay= merged.loc[indexList,['playlist_name','artist_name','track_name']]
dfDisplay


```
