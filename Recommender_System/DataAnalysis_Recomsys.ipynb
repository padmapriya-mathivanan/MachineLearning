```
from collections import Counter 
print(Counter(df['album_name'].values))
print(Counter(df['track_name'].values))
```

```
print(Counter(df['album_name'].values).most_common(5))
print(Counter(df['track_name'].values).most_common(5))
```

```
import matplotlib.pyplot as plt
bar_plot = dict(Counter(df['album_name'].values).most_common(5))
plt.bar(*zip(*bar_plot.items()))
plt.show()
```

```
import matplotlib.pyplot as plt
bar_plot = dict(Counter(df['track_name'].values).most_common(5))
plt.bar(*zip(*bar_plot.items()))
plt.show()
```

```
print(Counter(df['artist_name'].values))
print(Counter(df['artist_name'].values).most_common(5))
```

```
def get_frequencies(column_name):
    print(Counter(df[column_name].values))
    print(dict(Counter(df[column_name].values).most_common(5)))
    bar_plot = dict(Counter(df[column_name].values).most_common(5))
    plt.bar(*zip(*bar_plot.items()))
    plt.show()
```

```
bar_plot = dict(Counter(df['artist_name'].values).most_common(5))
plt.bar(*zip(*bar_plot.items()))
plt.show()
```

```
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from gensim.models import Word2Vec
from matplotlib import rcParams
from matplotlib.patches import Rectangle
from scipy.stats import relfreq
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA, LatentDirichletAllocation, KernelPCA
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from wordcloud import WordCloud

warnings.filterwarnings('ignore')

sns.set(context='poster', style='white')
rcParams['font.family'] = 'arial'
rcParams['font.sans-serif'] = 'arial'
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```
axes = df.drop(['pid', 'collaborative', 'modified_at', 'num_tracks', 'num_albums',
                'num_followers', 'num_edits','duration_ms',
       'num_tracks.1', 'pos', 'artist_name', 'track_uri', 'artist_uri',
       'track_name', 'album_uri', 'duration_ms.1', 'album_name'], axis=1)\
         .hist(figsize=(16,12), bins=20, alpha=0.75);
axes = axes.ravel()
plt.tight_layout()

titles = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'type', 'duration_ms.2','time_signature']
for i in range(len(axes)-1):
    axes[i].set_title(titles[i])
    if i % 4 == 0:
        axes[i].set_ylabel('Number of Songs')
plt.suptitle('Distributions of Spotify Audio Features', fontsize=30, y=1.05);
```

```
columns = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'type', 'duration_ms.2','time_signature']
df_scatter = df[columns]
sns.clustermap(df_scatter.corr(), cmap='PiYG', annot=True, figsize=(16,16))
plt.suptitle('Correlation Matrix Heatmap', fontsize=24, y=0.91)
```
